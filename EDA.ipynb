##### MODEL TUNING

##### Gradient Boosting Regressor

GradienBoosting Regressor:
Working:
1.	Initialization: Start with an initial prediction for all data points, typically the mean of the target values.
2.	Compute Residuals: Calculate the residuals, which are the differences between the actual target values and the current predictions.
3.	Fit Weak Learner: Train a weak learner (such as a small decision tree) on the residuals to identify patterns in the errors.
4.	Update Predictions: Adjust the current predictions by adding a fraction (controlled by the learning rate) of the weak learner’s predictions to the existing predictions.
5.	Iterate: Repeat the process of computing residuals, fitting weak learners, and updating predictions for a specified number of iterations or until the residuals are minimized, improving the model’s accuracy incrementally.

HYPERPARAMETERS:

Hyperparameters are settings or configurations that you specify before training a machine learning model. They control the training process and the structure of the model itself, and they are not learned from the data. Instead, they must be set prior to the training phase. The choice of hyperparameters can significantly impact the performance and effectiveness of the model.

Some of the Default parameters(hyper parameters):

n_estimators=100: The number of boosting stages to be run, i.e., the number of trees.

max_leaf_nodes=None: The maximum number of leaf nodes in the tree.

min_samples_split=2: The minimum number of samples required to split an internal node.

min_samples_leaf=1: The minimum number of samples required to be at a leaf node.

<b>Accuracy</b> : measures the proportion of correct predictions made by the model out of all predictions.<br>

<b>Confusion Matrix</b> : The confusion matrix provides a detailed breakdown of the model's performance, allowing us to identify specific types of errors.

<b>Precision & Recall</b> : Precision measures the proportion of true positive predictions among all positive predictions, while recall measures the proportion of true positive predictions among all actual positive instances.

<b>ROC</b> : AU-ROC represents the model's ability to discriminate between positive and negative classes. A higher AU-ROC value indicates better classification performance.

#### Decision tree 

##### Hyperparameters:
##### max_depth
Description: The maximum depth of the tree.
Default: None (nodes are expanded until all leaves are pure or until all leaves contain fewer than min_samples_split samples).

##### min_samples_split
Description: The minimum number of samples required to split an internal node.
Default: 2
Type: Integer or float. If float, it represents a fraction of the total number of samples.

##### min_samples_leaf
Description: The minimum number of samples required to be at a leaf node.
Default: 1
Type: Integer or float. If float, it represents a fraction of the total number of samples.
